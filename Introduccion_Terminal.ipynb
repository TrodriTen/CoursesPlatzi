# Introduccion a la terminal

- **Que es la terminal:**
  + Es una interfaz grafica que simula la linea de comandos
  + Shell - es la funcionalidad de la Termin al, toma comandos y se los da al SO para ejecutarlos
  + Un comando es un programa/instruccion que se ejecuta en la shell
  + La terminal es:
    * Flexible
    * Veloz
- **Que es un comando:**
  + Es un programa ejecutable - Funciones de Shell

- **Tipos Archivos:**
  + Al ejecutar el comando ls, parace lo siguente al inicio de la linea:
    1. "-" Este guion indica que el archivo listado es tipo de Archivo normal
    2. "d" Esta d indica que el archivo listado es un directorio
    3. "l" Esta l indica que el archivo listado es un link simbolico
    4. "b" Esta b indica que el archivo listado es de tipo especial (raro)
- **Permisos:**
  + Dueño (rwx)
  + Grupo (r-x)
  + World (r-x)
  + Significado de rwx:
    * r indica read, puede ser denotado con 0 o 1
    * w indica write, puede ser denotado con 0,1
    * x indica execute, puede ser denotado con 0,1
  + Configuraciones: Estan dadas en Binario.
    * rwx ---- 111 = 7
    * rw- ---- 110 = 6
    * r-x ---- 101 = 5
    * r-- ---- 100 = 4
    * -w- ---- 010 = 2
    * --x ---- 001 = 1


- **Modelo Simbolico:**
  + Asignar permisos en la terminal a archivos con chmod/chowner
  + u --- usuarios
  + g --- grupo
  + o --- otros (world)
  + a --- aplicar a todos

## Comandos Linux

- **pwd** (Saber el directorio actual)
- **cd** "nombre carpeta" (Cambiar de directorio)
- **cd ..** (Devolverse de directorio)
- **ls** (listar directorios y archivos del directorio actual)
- **clear** (Limpiar terminal)
- **reset** (Reiniciar la terminal)
- **touch "nombre archivo.extension"** (Crear un archivo o varios en el directorio actual)
- **code .** (Abrir Vscode en el directorio actual)
- **mkdir "nombre directorio"** (Crear un nuevo o varios directorios)
- **rm** (Eliminar archivos y/o directorios de forma definitiva)

- **mv "nombre archivo" "nuevo nombre"** (Renombrar un item)
- **cat "nombre archivo"** (Mostrar contenido de un archivo en la terminal y/o concatenar archivos)
- **grep "palabra"** (Buscar texto en archivos o comandos)
- **ps** (Muestra los procesos en ejecucion)
- **kill "ID proceso"** (Matar/Detener un proceso)
- **chmod "numero binario" "nombre archivo"** (Cambiar permisos de archivos o directorios)
- **chmod u + "permiso"** (Da permiso a usuarios)
- **chmod u - "permiso"** (Quita permiso a usuarios)
- **sudo** (Ejecucion de comandos en SuperUsuario)
- **ls -a** (Listar los archivos y directorios junto con elementos ocultos)
- **ls -l** (Listado detallado de los archivos y directorios)
- **ls -lh** (Listar archivos y directorios para la lecturta humana)
- **rmdir "directorio"** (Eliminar directorio)
- **cd ../..** (Devolverse a dos directorios arriba)
- **file "nombre archivo"** (Da la informacion del archivo en cuestion)
